## Определяющие уравнения 


Пусть геологическая среда описывается в акустическом приближении, в котором существуют только продольные волны. Тогда они характеризуются одним параметром, зависящим от радиус-вектора и называемым \textit{скоростью распространения акустической волны с}. Если ввести в расмотрение величину, равную давлению в среде $P(\textbf{r}, t)$, то она удовлетворяет уравнению: 
\begin{equation} 
\nabla^2 P(\textbf{r},t) - \frac{1}{c^2(\textbf{r})}\frac{\partial^2}{\partial t^2} P(\textbf{r},t) = - F^e (\textbf{r}, t). 
\end{equation} 

Решение этого уравнения в случае, когда $F^e (\textbf{r}, t)$ - точечный источник (дельа-функция по времени и по пространству), называется \textit{функцией Грина}, и для однородного бесконечного пространства она может быть записана в виде 
\begin{equation} 
G^w (\textbf{r},t|\textbf{r}', t') = \frac{1}{4\pi |\textbf{r} - \textbf{r}'|}\delta (t - t' - \frac{|\textbf{r} - \textbf{r}'|}{c}) 
\end{equation} 
где $\textbf{r}', t'$ - место расположения источника и время, в которое он начал действовать. 


## Формула Кирхгофа 


В общем случае как для ограниченной, так и для неограниченной области можно записать интегральную связь между полем давления на её границе $S$  и полем в произвольной точке внутри неё. Эта связь называется формулой Кирхгофа и представляется в виде 
\begin{equation} 
P(\textbf{r}', t') = \int_S \int_{-\infty}^{+\infty}[G^w(\textbf{r}', t'|\textbf{r}, t) \frac{\partial}{\partial n}P(\textbf{r}, t) - P(\textbf{r}, t)\frac{\partial}{\partial n} G^w(\textbf{r}', t'|\textbf{r}, t)]dtds. \end{equation} 

Здесь штриховые координаты соответствуют точке, лежащей внутри рассматриваемой области. Подставляя в (3) выражение для формулы Грина в явном виде (2), получим 
\begin{equation} 
P(\textbf{r}', t') = \frac{1}{4\pi}\int_{S} [\frac{1}{c|\textbf{r}' - \textbf{r}|} \frac{\partial |\textbf{r}' - \textbf{r}|}{\partial n} \frac{\partial P'}{\partial t'} - \frac{\partial}{\partial n}(\frac{1}{|\textbf{r}' - \textbf{r}|})P' + \frac{1}{|\textbf{r}' - \textbf{r}|}\frac{\partial}{\partial n} P']ds, 
\end{equation} 
где $P' = P(\textbf{r}, t' - \frac{|\textbf{r}' - \textbf{r}|}{c})$. 

Рассмотрим случай однородного полупространства, ограниченного сверху плоскостью z = 0 (стандартные площадные полевые сейсморазведочные работы). В этом случае $\frac{\partial}{\partial z}$, следовательно выражение (4) 
может быть упрощено до вида 
\begin{equation} 
P(\textbf{r}', t') = \frac{1}{4\pi}\int_S [\frac{z' - z}{|\textbf{r}' - \textbf{r}|}(\frac{1}{c} \frac{\partial P'}{\partial t'} + \frac{1}{|\textbf{r}' - \textbf{r}|^2} P') - \frac{1}{|\textbf{r}' - \textbf{r}|} \frac{\partial P'}{\partial z}]ds. 
\end{equation} 

### Импульс Риккера
Разберемся теперь со временной функцией источника. Пусть на поверхность полупространства воздействует давление в виде импульса Риккера: 
\begin{equation} 
P_{ricker} = P_0(1 - 2\pi^2 f^2 t^2) e^{-\pi^2 f^2 t^2}, 
\end{equation} 
тогда функция $P'(\textbf{r}', t')$ может быть записана в виде 
\begin{equation} 
P'(\textbf{r}', t') = P(\textbf{r}, t' - \frac{|\textbf{r}' - \textbf{r}|}{c}) = P_0(1 - 2\pi^2 f^2 (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})^2) e^{-\pi^2 f^2 (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})^2} 
\end{equation}. 

С 

учётом (7) аналитически могут быть получены значения 
\begin{equation} 
\frac{\partial P'}{\partial t'} = -2\pi^2 f^2 P_0(3 - 2\pi^2 f^2 (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})^2) e^{-\pi^2 f^2 (t' - \frac{\textbf{r}' - \textbf{r}}{c})^2} 
\end{equation} 
и 
\begin{equation} 
\frac{\partial P'}{\partial z} = -2\pi^2 f^2 P_0 (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})[3 - 2\pi^2 f^2 (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})^2] e^{-\pi^2 f^2 (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})^2} \frac{z' - z}{c|\textbf{r}' - \textbf{r}|}. 
\end{equation} 
Таким образом, может быть рассчитано поле в произвольной точке внутри области, ограниченной поверхностью S. 
\newline 

Рассчитаем поле в произвольной точке внутри области для различных временных зависимостей
сигнала-источника: Берлаге, Пузырева, Гельфанда, затухающей синусоиды, колокольной огибающей.

### Импульс Пузырева

Пусть теперь на поверхность полупространства воздействует давление в виде импульса Пузырева: 
$$ 
P_{puz} = P_0 e^{-\beta^2 t^2} cos(f t), 
$$ 
тогда функция $P'(\textbf{r}', t')$ будет записана в виде 
$$ 
P'(\textbf{r}', t') = P_0 e^{-\beta^2 (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})^2}cos(f(t' - \frac{|\textbf{r}' - \textbf{r}|}{c})) 
$$ 
Получим частные производные
$$
\frac{\partial P'}{\partial t'} = -P_0 e^{-\beta^2 (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})^2}(2\beta^2 (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})cos(f(t' - \frac{|\textbf{r}' - \textbf{r}|}{c})) + f sin(f(t' - \frac{|\textbf{r}' - \textbf{r}|}{c})))
$$
и 
$$ 
\frac{\partial P'}{\partial z} = P_0 (\frac{r'_2 - z}{c|\textbf{r}' - \textbf{r}|}) e^{-\beta^2 (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})^2}( 2\beta^2 (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})cos(f(t' - \frac{|\textbf{r}' - \textbf{r}|}{c})) + f sin(f(t' - \frac{|\textbf{r}' - \textbf{r}|}{c}))) 
$$

### Импульс Берлаге

Пусть теперь на поверхность полупространства воздействует давление в виде импульса Пузырева: 
$$ 
P_{ber} = P_0 t^{\alpha} e^{-\beta t} \sin(f t), 
$$ 
тогда функция $P'(\textbf{r}', t')$ будет записана в виде 
$$ 
P'(\textbf{r}', t') = P_0 (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})^{\alpha} e^{-\beta (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})} \sin(f (t' - \frac{|\textbf{r}' - \textbf{r}|}{c})), 
$$ 
Получим частные производные 
$$ 
\frac{\partial P'}{\partial t'} = P_0 \tau^{\alpha}e^{-\beta \tau} (\alpha \frac{\sin (f \tau)}{\tau} - \beta \sin (f \tau) + f\cos (f\tau))
$$ 
и 
$$ 
\frac{\partial P'}{\partial z} = P_0 \tau^{\alpha}(\frac{r'_2 - z}{c|\textbf{r}' - \textbf{r}|})e^{-\beta \tau} (\alpha \frac{\sin (f \tau)}{\tau} - \beta \sin (f \tau) + f\cos (f\tau))
$$

# Импульс Риккера

```python
def pressure_ricker (r_0, r, t_0): #давление от источника
    v = - (np.pi**2) * (f**2)
    mod = np.sqrt ((r_0[0] - r[0])**2 + (r_0[1] - r[1])**2 + (r_0[2] - r[2])**2)
    res = P_0 * (1 + 2*v * (t_0 - mod/c)**2)* np.exp (v * (t_0 - mod/c)**2)
    return res

def partial_t_ricker (r_0, r, t_0): # частная производная P' по t
    v = - (np.pi**2) * (f**2)
    
    mod = np.sqrt ((r_0[0] - r[0])**2 + (r_0[1] - r[1])**2 + (r_0[2] - r[2])**2)
    
    res = 2*v * P_0 * (3 + 2*v * (t_0 - mod/c)**2)* np.exp (v * (t_0 - mod/c)**2)
    return res

def partial_z_ricker(r_0, r, t_0): #r_0 -- точка, в которой считаем давление, r -- радиус вектор точки источника, t_0 -- момент времени, в который измеряется давление в точке
    r_0 = np.array(r_0)
    r = np.array(r)
    res = - 2 * np.pi**2 * f**2 * P_0
    res*= (t_0 - length(r-r_0)/c)
    res*= (3-2 * np.pi**2 * f**2 *(t_0 - length(r-r_0)/c)**2)
    res*= np.exp(- np.pi**2 * f**2 * (t_0 - length(r-r_0)/c)**2)
    res*= (r_0[2]-r[2])/(c*length(r-r_0))
    return res
```

# Импульс Берлаге

```python
def pressure_berlage(r_0, r, t_0): #r_0 -- точка, в которой считаем давление, r -- радиус вектор точки источника, t_0 -- момент времени, в который измеряется давление в точке
    tau = t_0 - length(r - r_0)/c
    press = P_0 * tau**alpha_berl * np.exp(-beta_berl * tau)*np.sin(f* tau)
    return press

def partial_t_berlage(r_0, r, t_0):
    tau = t_0 - length(r - r_0)/c
    if tau <= 0:
        return 0
    res = P_0 * tau ** alpha_berl * np.exp(-beta_berl * tau)*(np.sin(f * tau) * (alpha_berl / tau - beta_berl) + f * np.cos(f * tau))
    return res

def partial_z_berlage(r_0, r, t_0):
    tau = t_0 - length(r - r_0)/c
    if tau <= 0:
        return 0
    res =  P_0 * tau ** alpha_berl * np.exp(-beta_berl * tau)*(np.sin(f * tau) * (alpha_berl / tau - beta_berl) + f * np.cos(f * tau)) * (r_0[2] - r[2])/(c * length(r - r_0))
    return res
```

# Импульс Пузырёв

```python
beta = 1 

def pressure_p (r_0, r, t_0): 
    mod = np.sqrt ((r_0[0] - r[0])**2 + (r_0[1] - r[1])**2 + (r_0[2] - r[2])**2)
    res = P_0 * np.exp (-(beta * (t_0 - mod/c))**2)*np.cos(f*(t_0 - mod/c)) 
    return res 

def partial_t_p (r_0, r, t_0): 
    mod = np.sqrt ((r_0[0] - r[0])**2 + (r_0[1] - r[1])**2 + (r_0[2] - r[2])**2)
    res = P_0*np.exp (-(beta * (t_0 - mod/c))**2)*(-2*beta**2*(t_0 - mod/c)*np.cos(f*(t_0 - mod/c)) - f*np.sin (f*(t_0 - mod/c))) 
    return res 

def partial_z_p (r_0, r, t_0): 
    mod = np.sqrt ((r_0[0] - r[0])**2 + (r_0[1] - r[1])**2 + (r_0[2] - r[2])**2)
    mod_p = (1/mod)*(r[2] - r_0[2]) 
    res = P_0 * np.exp(-(beta * (t_0 - mod/c))**2)*(beta**2/c*2*(t_0 - mod/c)*mod_p*np.cos(f*(t_0 - mod/c)) +f/c*np.sin(f*(t_0 - mod/c))*mod_p) 
    return res   
```

# Функция Кирхгофа

```python
def integrand (r_0, r, t_0, pressure, partial_t, partial_z): #подынтегральная функция
    mod = np.sqrt ((r_0[0] - r[0])**2 + (r_0[1] - r[1])**2 + (r_0[2] - r[2])**2)
    
    res =  (r_0[2] - r[2])/mod *((partial_t (r_0, r, t_0) / c + pressure (r_0, r, t_0) / mod**2 ) - partial_z (r_0, r, t_0)/ mod)
    
    return res
```

# Функция определения области поверхности, от которой дошло возмущение

```python
def g(x, y, r_0, t_0): # функция задаёт множество по которому происходит интегрирование
    r = np.array([x, y, 0])
    tau = t_0 - abs(length(r - r_0))/c
    if tau <= 0:
        return -1
    else:
        return 1
```

# Интегрирование методом прямоугольников

```python
def rectangular_double1(g, x0, x1, y0, y1, n, pressure, partial_t, partial_z, r_0, t_0): #метод прямоугольников
    #Вычисляет значение двойного интеграла по формуле прямоугольников
    #при реализации двойной суммы
    #f - подынтегральная функция
    #a, b, c, d - границы прямоугольной области интегрирования
    #nx, ny - количество частичных отрезков по x и y соответственно
    hx = abs(x1 - x0)/float(n)
    hy = abs(y1 - y0)/float(n)
    I = 0
    for i in range(n):
        for j in range(n):
            xi = x0 + hx/2 + i*hx
            yj = y0 + hy/2 + j*hy
            if g(xi, yj, r_0, t_0) < 0:
                break
            r = [xi, yj, 0]
            add = hx*hy*integrand (r_0, r, t_0, pressure, partial_t, partial_z)
            I+=add
    return I
```

# Интегрирование методом Монте-Карло

```python
def MonteCarlo_double(g, x0, x1, y0, y1, n, pressure, partial_t, partial_z, r_0, t_0):
    #Численное интегрирование методом Монте-Карло функции
    #f по области g >= 0, помещенной в прямоугольник
    #[x0, x1]x[y0, y1].
    #n^2 - количество случайных точек
    x = np.random.uniform(x0, x1, n)
    y = np.random.uniform(y0, y1, n)
    f_mean = 0          # среднее значение f
    num_inside = 0      # число точек, попавших в область интегрирования (g>=0)
    for i in range(len(x)):
        for j in range(len(y)):
            if(g(x[i], y[j], r_0, t_0) > 0):
                num_inside += 1
                f_mean += integrand (r_0, [x[i], y[j],0], t_0, pressure, partial_t, partial_z)

    f_mean = f_mean/float(num_inside)
    area = num_inside/float(n**2)*(x1 - x0)*(y1 - y0)
    return area*f_mean
```

# Интегрирование методом трапеций

```python
def trapezium_2d_method(g, x0, x1, y0, y1, n, pressure, partial_t, partial_z, r_0, t_0): #метод прямоугольников
    #Вычисляет значение двойного интеграла по формуле прямоугольников
    #при реализации двойной суммы
    #f - подынтегральная функция
    #a, b, c, d - границы прямоугольной области интегрирования
    #nx, ny - количество частичных отрезков по x и y соответственно
    hx = abs(x1 - x0)/float(n)
    hy = abs(y1 - y0)/float(n)
    I = 0
    for i in range(n):
        for j in range(n):
            xi = x0 + hx/2 + i*hx
            yj = y0 + hy/2 + j*hy
            if g(xi, yj, r_0, t_0) < 0:
                break
            q = 1
            if (i == 0) or (i == n - 1):
                q*=1./2
            if (j == 0) or (j == n - 1):
                q*=1./2
            r = [xi, yj, 0]
            I+= hx*hy*q*integrand(r_0, r, t_0, pressure, partial_t, partial_z)
    return I
```

# Функция расчёта давления в узлах сетки

```python
def get_pressure_map(x_0, x_1, y, z_1, t_1, Nx, Nz, Nt, pressure, partial_t, partial_z, integral): #Функция возвращает сетку распределения давления pressure  в сечении [x_0,x_1]x[y]x[0,z_1]x[0,t_1], с шагом stepx, stepz, stept  
    res = np.zeros((Nt, Nz, Nx))
    stepx = abs(x_1 - x_0)/Nx
    stepz = abs(z_1)/Nz
    stept = abs(t_1)/Nt
    nt = 0
    for t in [stept * i for i in range(Nt)]:
        nx = 0
        for x in [x_0 + stepx * i for i in range(Nx)]:
            nz = 0
            for z in [stepz * i for i in range(Nz)]:
                gx = x_0
                while g(gx, y, np.array([x, y, -z]), t) > 0:
                    gx += stepx
                delta = abs(x_0 - gx)
                res[nt, nz, nx] = integral(g, x_0 - delta, x_0 + delta, y - delta, y + delta, Nx, pressure, partial_t, partial_z, np.array([x, y, -z]), t)
                print 'Обработан узел ' + str(nt)+ ' ' + str(nx) + ' ' + str(nz)
                nz += 1
            nx += 1
        nt += 1
    return res
```

```python
def get_pressure_map_1d(z_1, t_1, Nz, Nt, pressure, partial_t, partial_z, integral, show): #Функция возвращает сетку распределения давления pressure  в сечении [x_0,x_1]x[y]x[0,z_1]x[0,t_1], с шагом stepx, stepz, stept  
    res = np.zeros((Nt, Nz))
    stepz = abs(z_1)/Nz
    stept = abs(t_1)/Nt
    nt = 0
    for t in [stept * i for i in range(Nt)]:
        nz = 0
        for z in [stepz * i for i in range(Nz)]:
            gx = 0
            while g(gx, y, np.array([0, 0, -z]), t) > 0:
                gx += stepz
            delta = abs(x_0 - gx)
            res[nt, nz] = integral(g,  - delta,  delta, - delta,  delta, Nx, pressure, partial_t, partial_z, np.array([0, 0, -z]), t)
            if (show > 0):
                print 'Обработан узел ' + str(nt)+ ' ' + ' ' + str(nz)
            nz += 1
        nt += 1
    return res
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

fig, ax = plt.subplots(nrows=Nt, ncols=1,sharex=True,figsize=[10, 10 * Nt]) 

z = np.linspace(0, -z_1, Nz)
stept = abs(t_1)/Nt

for i in range(Nt):
    plt.subplot( Nt,1, i+1)
    plt.scatter(z, abs(dat[i]), color = 'red',s=10, marker='o')
    plt.xlabel(str(i*stept))
    ax[i].axes.set_aspect('equal')
    extent = ax[i].get_window_extent().transformed(fig.dpi_scale_trans.inverted())
    fig.savefig('berlage_' + str(i) + '_1_d.png', bbox_inches=extent.expanded(1.2, 1.2))


fig.tight_layout()
```
